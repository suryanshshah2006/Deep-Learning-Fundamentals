With random_state=1 (or any fixed number, like 42): The function uses that number (1) as a seed for its internal random number generator. Think of the seed as setting the starting point for the shuffle.

Every time you run the code with random_state=1, the deck will be shuffled in the exact same way.

Therefore, the exact same customers will end up in the Training Pile and the Testing Pile every single time.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state=1)
```

```python
x_train.shape, x_test.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

input_dim=60 is equal to input.shape(60,)

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

The batch_size=8 parameter determines how many training samples the neural network processes at once before updating its internal parameters (weights and biases).

```python
model.fit(x_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(x_test, y_test)
```

so this clearly shows overfitting

```python
y_pred = model.predict(x_test).reshape(-1)
print(y_pred[:10])
```

```python
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test, y_pred))
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred)
cm
```

Code with dropout regularization

```python
modeld = keras.Sequential([
    keras.layers.Dense(60, input_dim=60, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])
```

keras.layers.Dropout(0.5), = drops 50 percent of the neurons

```python
modeld.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
modeld.fit(x_train, y_train, epochs=100, batch_size=8)
```

```python
modeld.evaluate(x_test, y_test)
```

even though there is much improvement in test dataset but that was the syntax and we have to hit and try with the dropout layers to get a good improvement

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = modeld.predict(x_test).reshape(-1)
y_pred = np.round(y_pred)
print(classification_report(y_test, y_pred))
```
